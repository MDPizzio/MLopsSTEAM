```python
#Se importa la librería a usar (posterior a la instalación del paquete TextBlob) para realizar el análisis de sentimientos
from textblob import TextBlob
```

```python
#Se crea la función para realizar el análisis de sentimientos
def sentiment_analysis(review):
    if isinstance(review, list) and len(review) > 0:
        text = review[0].get('review', '')
        sentiment = TextBlob(text).sentiment.polarity
        
        if sentiment < -0.2:
            return 0 # Negativo
        elif sentiment >= -0.2 and sentiment <= 0.2:
            return 1 # Neutral
        else:
            return 2 # Positivo
    else:
        return 1 # Valor para casos en que la reseña escrita este ausente
```

```python
df_reviews['sentiment_analysis'] = df_reviews['reviews'].apply(sentiment_analysis)
```

```python

   
from datetime import datetime
import calendar
import re
import pandas as pd

def convert_date(date_str):
    if pd.notna(date_str) and isinstance(date_str, str):
        # Usamos expresiones regulares para extraer el día, mes y año
        match = re.match(r'Posted (\w+) (\d+), (\d+)', date_str)
        if match:
            month_name, day, year = match.groups()

            # Convertimos el nombre del mes a número usando el módulo calendar
            month_number = list(calendar.month_abbr).index(month_name[:3])

            # Creamos la nueva fecha en el formato deseado 'yyyy-mm-dd'
            new_date = datetime(int(year), month_number, int(day))
            return new_date.strftime('%Y-%m-%d')
    return pd.NaT  # Devolvemos nulo si no coincide el formato

```

```python
from datetime import datetime
import calendar
import re
import pandas as pd

def extract_year(date_str):
    if pd.notna(date_str) and isinstance(date_str, str):
        # Usamos expresiones regulares para extraer el año
        match = re.match(r'Posted .* (\d{4})', date_str)
        if match:
            year = match.group(1)
            return int(year)
    return pd.NaT  # Devolvemos nulo si no coincide el formato
```

```python
df_reviews_aux2['posted'] = df_reviews_aux2['posted'].apply(extract_year)
df_reviews_aux2
```
